This is a project carried out for learning purposes only. The conclusions made are based on my understanding of the dataset used, thus they do not have any real life statistical implication as this is for learning purposes.


This is an  analytical project carried out based on the data collected from `WeRateDogs @dog_rates`, a twitter account that rates different dogs on a defined scale of values with denominator 10.Three files were collected: `image recognition file` which contains images of dogs that have been passed through image recognition system, `archive files` and the `tweet files`. These data sets contains dog ratings from 2015 to 2017 with columns but not limited to: dog name, tweet_id, retweet_count, favorite_count, image recognition confidence levels and a lot of other descriptive columns.

The data cleaning proces was a tedius task as lots of issues were observed and fixed ranging from missing values to renaming columns and merging files. The cleaning process is the most time consuming and difficult part of the data wrangling process for a Data Analyst.

The following insights were made from the final cleaned data:
1. The highest number of tweets for rated dogs was in 2016. This was the second year of the twitter account existence, assuming 2015 was the starting year. This can be seen in the viaulaization below. The year 2016 had 974 tweet ratings as compared with 2015 and 2017 with 663 and 357 tweets respectively.





2. From the cleaned dataset, it was observed that a dog named Atticus had the highest rating of 1776, even though this rating is outrageously high, investigations shoed it is correct. The tweet_id is 749981277374128128, and this rating was in 2016. An issue here is that the machine learning algorithms used flagged the image as not a dog.

3. The dogs with the lowest ratings have no name. This rating is diasatrous because it is zero(0). Their tweet_ids are 746906459439529985 and 835152434251116546 with both tweets made in 2016 and 2017 respectively.

4. It was obserced that most tweets, 98% approximately was made from an iphone, which tells that most twitter users that visit this page (@dog_rates) makes use of the apple iphone.


5. The image prediction models accuracy was put to the test to determine which is better among the three algorithms: `p1`, `p2`, and `p3`. p2 came out tops with `1495` True predictions, the most True predictions in all as shown in the visualization below.


6. Algorithm p3 had the worst prefiction with 548 False image predictions among the three.
